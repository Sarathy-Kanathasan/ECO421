```python
import numpy as np
import pandas as pd
df= pd.read_csv("traffic_crashes.csv")
# df.head()
# df.info()

#notice there are several columns will several null elements, Its better to remove these columns then to lower
#the test and training count
df2 = df[df.columns[df.isnull().mean() < .60]]
# df2.info()

df2.dropna(inplace= True)
# df2.head()
df3 = df2

# data["bank_acct_new"] = data.bank_acct.replace({"yes": 1, "no": 0})
#standardize quantitative values
# # first_crash = df3['FIRST_CRASH_TYPE'].unique()
# tcd = df3['TRAFFIC_CONTROL_DEVICE'].unique()
# tcd
# tcd_dict = {}
# i = 0
# for type  in tcd:
#     tcd[type] = i
#     i = i+1
#  clean_qual =  {'FIRST_CRASH_TYPE':     {'TURNING': 0, 'ANGLE': 1,'REAR END': 2, 'SIDESWIPE SAME DIRECTION': 3,
#                                          'FIXED OBJECT': 4,'SIDESWIPE OPPOSITE DIRECTION': 5,'PARKED MOTOR VEHICLE': 6,
#                                          'PEDESTRIAN': 7, 'PEDALCYCLIST': 8, 'OTHER OBJECT': 9, 'HEAD ON': 10, 
#                                          'OTHER NONCOLLISION': 11, 'OVERTURNED': 12, 'ANIMAL': 13, 'REAR TO SIDE': 14, 
#                                          'TRAIN': 15,'REAR TO FRONT': 16},
                
#                 "num_cylinders": {"four": 4, "six": 6, "five": 5, "eight": 8,
#                                   "two": 2, "twelve": 12, "three":3 }
#                }

df3.info()
```

```python
#standardize quantitative values
# first_crash = df3['FIRST_CRASH_TYPE'].unique()

#encode Traffic control device 
tcd = df3['TRAFFIC_CONTROL_DEVICE'].unique()
tcd
tcd_dict = {}
i = 0
for type  in tcd:
    tcd_dict[type] = i
    i = i+1

#encode device conditions
dcc = df3['DEVICE_CONDITION'].unique()
dcc
dcc_dict = {}
i = 0
for type  in dcc:
    dcc_dict[type] = i
    i = i+1
# dcc_dict

#encode weather conditions
wc = df3['WEATHER_CONDITION'].unique()
wc
wc_dict = {}
i = 0
for type  in wc:
    wc_dict[type] = i
    i = i+1
# wc_dict

#Encode lighting Conditions
lc = df3['LIGHTING_CONDITION'].unique()
lc
lc_dict = {}
i = 0
for type  in lc:
    lc_dict[type] = i
    i = i+1
# lc_dict

#Encoding crash type
fct = df3['FIRST_CRASH_TYPE'].unique()
fct
fct_dict = {}
i = 0
for type  in fct:
    fct_dict[type] = i
    i = i+1
# fct_dict

#Encode Traffic type
tt = df3['TRAFFICWAY_TYPE'].unique()
tt
tt_dict = {}
i = 0
for type  in tt:
    tt_dict[type] = i
    i = i+1
tt_dict

#Encode ALIGNMENT type
a = df3['ALIGNMENT'].unique()
a
a_dict = {}
i = 0
for type  in a:
    a_dict[type] = i
    i = i+1
a_dict

#encode ROADWAY_SURFACE_COND  type
rsc = df3['ROADWAY_SURFACE_COND'].unique()
rsc
rsc_dict = {}
i = 0
for type  in rsc:
    rsc_dict[type] = i
    i = i+1
rsc_dict

#encode ROAD_DEFECT type
rd = df3['ROAD_DEFECT'].unique()
rd
rd_dict = {}
i = 0
for type  in rd:
    rd_dict[type] = i
    i = i+1
rd_dict

#encode REPORT_TYPE type
rt = df3['REPORT_TYPE'].unique()
rt
rt_dict = {}
i = 0
for type  in rt:
    rt_dict[type] = i
    i = i+1
rt_dict

# Encode CRASH_TYPE type
ct = df3['CRASH_TYPE'].unique()
ct
ct_dict = {}
i = 0
for type  in ct:
    ct_dict[type] = i
    i = i+1
ct_dict

#encode DAMAGE type
d = df3['DAMAGE'].unique()
d
d_dict = {}
i = 0
for type  in d:
    d_dict[type] = i
    i = i+1
d_dict

#encode STREET_DIRECTION 
dpn = df3['STREET_DIRECTION'].unique()
dpn
dpn_dict = {}
i = 0
for type  in dpn:
    dpn_dict[type] = i
    i = i+1
dpn_dict

#encode PRIM_CONTRIBUTORY_CAUSE
pcc = df3['PRIM_CONTRIBUTORY_CAUSE'].unique()
pcc
pcc_dict = {}
i = 0
for type  in pcc:
    pcc_dict[type] = i
    i = i+1
pcc_dict

#encode for MOST_SEVERE_INJURY 
msi = df3['MOST_SEVERE_INJURY'].unique()
msi
msi_dict = {}
i = 0
for type  in msi:
    msi_dict[type] = i
    i = i+1
msi_dict

df3.head()
```

```python
cleanup_nums = {'TRAFFIC_CONTROL_DEVICE':  tcd_dict   ,
                'DEVICE_CONDITION': dcc_dict,
                'WEATHER_CONDITION':wc_dict,
                'LIGHTING_CONDITION':lc_dict,
                'FIRST_CRASH_TYPE':fct_dict,
                'TRAFFICWAY_TYPE':tt_dict,
                'ALIGNMENT':a_dict,
                'ROADWAY_SURFACE_COND':rsc_dict,
                'ROAD_DEFECT':rd_dict,
                'REPORT_TYPE':rt_dict,
                'CRASH_TYPE':ct_dict,
                'DAMAGE':d_dict,
                'STREET_DIRECTION':dpn_dict,
                'PRIM_CONTRIBUTORY_CAUSE':pcc_dict,
                'MOST_SEVERE_INJURY':msi_dict
                }
df4 = df3.replace(cleanup_nums)

#remove unecessary features
del df4['CRASH_RECORD_ID']
del df4['RD_NO']
del df4['CRASH_DATE']
del df4['DATE_POLICE_NOTIFIED']
del df4['SEC_CONTRIBUTORY_CAUSE']
del df4['STREET_NAME']
del df4['LOCATION']

df4.head()
```
